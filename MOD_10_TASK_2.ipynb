```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Utwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10,000 próbek
data_100 = np.random.normal(loc=3, scale=1, size=100)
data_10000 = np.random.normal(loc=3, scale=1, size=10000)

# Przekształć dane na obiekty DataFrame
df_100 = pd.DataFrame(data_100, columns=['Value'])
df_10000 = pd.DataFrame(data_10000, columns=['Value'])

# Oblicz statystyki opisowe
stats_100 = df_100.describe()
stats_10000 = df_10000.describe()

# Wyświetl statystyki
print("Statystyki dla 100 próbek:")
print(stats_100)

print("\nStatystyki dla 10,000 próbek:")
print(stats_10000)
```

```python
dff = data_100
plt.hist(dff, bins=20)

mean_dff = np.round(np.mean(dff))
print(mean_dff)
plt.axvline(x = mean_dff, color='red', label = 'Średnia')

median_dff = np.median(dff)
print(median_dff)
plt.axvline(x = median_dff, color='green', label = 'Mediana')


mode_dff = stats.mode(dff)
print(mode_dff)
plt.axvline(x = mode_dff[0], color='orange', label = 'Moda')

q0 = np.quantile(dff, 0.0)
q1 = np.round(np.quantile(dff, 0.25))
q3 = np.round(np.quantile(dff, 0.75))
q4 = np.round(np.quantile(dff, 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_dff}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')

range_dff = max(dff) - min(dff)
print('range:', range_dff)
iqr = q3 - q1
print('iqr:', iqr)
print('kurtosis',
    stats.kurtosis(dff, bias=False), pd.DataFrame(dff).kurtosis()[0],
    'skew',
    stats.skew(dff, bias=False), pd.DataFrame(dff).skew()[0],
    sep='\n'
)

plt.legend(loc='upper right')
plt.show()
```
